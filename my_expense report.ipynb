```python
Trans_count = df["Category"].value_counts()
Trans_count
```

```python
Trans_perc = round(df["Category"].value_counts(normalize = True)*100,2)
Trans_perc
```

```python
category_grp = df.groupby(["Category"])
card_grp = df.groupby(["Card Type"])
year_grp = df.groupby(["year"])
amt_sum_card = card_grp["Amount"].sum()
amt_sum_year = year_grp["Amount"].sum()
amt_sum_category = category_grp["Amount"].sum()
card_list = sorted(list(df["Card Type"].unique()))
year_list = sorted(list(df["year"].unique()))
category_list = sorted(list(df["Category"].unique()))
```

```python
plt.barh(card_list,amt_sum_card)
plt.title("Amount by Card Type")
plt.xlabel("Amount in USD($)")
plt.show()
```

```python
plt.bar(year_list,amt_sum_year)
plt.title("Amount by Year")
plt.ylabel("Amount in USD($)")
plt.show()
```

```python
plt.barh(category_list,amt_sum_category)
plt.title("Amount by Category")
plt.xlabel("Amount in USD($)")
plt.show()
```
